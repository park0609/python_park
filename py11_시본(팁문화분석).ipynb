```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1) #1행 2열중 첫번째 뜻 
ax2 = fig.add_subplot(1,2,2) #1행 2열중 두번째 뜻

sns.stripplot(x = "day" , y = "tip" , hue="sex" , data=tips, alpha = 0.5 , ax = ax1)
ax1.set_title("stripplot으로 표현한 날짜별로 분류한 성별과의 팁관계")

sns.swarmplot(x = "day" , y = "tip" , hue="sex" , data=tips, palette= "Set2",alpha = 0.5 , ax = ax2)
ax2.set_title("swarmplot으로 표현한 날짜별로 분류한 성별과의 팁관계")


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1) #1행 2열중 첫번째 뜻 
ax2 = fig.add_subplot(1,2,2) #1행 2열중 두번째 뜻

sns.regplot(x = "total_bill" , y = "tip", data=tips, color="blue", scatter_kws={"s":25,"alpha":0.5},line_kws={"linestyle":"--"},ax=ax1)
sns.regplot(x = "total_bill" , y = "tip", data=tips, color="blue", scatter_kws={"s":25,"alpha":0.5},fit_reg=False, ax=ax2)

#선형회귀란? 어떤 변수가 특정변수의 값에 영향을 주는것을 것을 말함, 변수가 x 특정변수가 y라 했을 때 x는 독립적으로 변할 수있지만 y는 x에 영향을 받아 값이 바뀜 한개면 단순 2개이상은 다중
fig.suptitle("Scatter plots 선형회귀의 ON/OFF 비교")
ax1.set_title("선형회귀 on")
ax2.set_title("선형회귀 off")


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
exercise = sns.load_dataset("exercise")
exercise.sample(5)
```

```python
plt.figure(figsize=(10,5))
sns.pointplot(x="time",y="pulse",hue="kind",data=exercise,capsize=0.1)
plt.title("시간별 심박수와 운동과의 관계")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="time",y="pulse",hue="kind",data=exercise, palette="Set2")
plt.title("시간별 심박수와 운동과의 관계")
plt.savefig("시간별 심박수와 운동과의 관계")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.swarmplot(x="time",y="pulse",hue="kind",data=exercise,dodge=True, palette="Set1" )
plt.title("시간과 운동 유형의 따른 개별 맥박 측정")
plt.savefig("시간과 운동 유형의 따른 개별 맥박 측정")
plt.show()
```
