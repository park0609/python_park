### 2024 통계

```python
file_path = "./ref/청소년정신건강2.csv"
ori = pd.read_csv(file_path)
```

```python
new_name = ['시점','스트레스','스트레스남학생','스트레스여학생','우울감경험율','우울남학생','우울여학생','자살생각율','자살남학생','자살여학생']
ori.columns = new_name
```

```python
ori = ori.loc[1:,"스트레스":]
ori
```

```python
ori.loc[2] = 100 - ori.loc[1,"스트레스":].astype(float)#astype은 형태를 전환시킬수있다. float int, str로도 바꿀수있다.
ori["응답"] = ["그렇다","아니다"]
ori
```

```python
ori = ori.set_index("응답") # 컬럼 하나를 인덱스로 지정, reset_index()는 원래 인덱스 값을 사용하겠다라는 기능
ori
```

```python
ori["스트레스"].astype(float).plot.pie(explode = [0,0.02])
```

### 24~25 통계

```python
fig , ax = plt.subplots(2,3,figsize = (10,8))

ori["스트레스"].astype(float).plot.pie(explode = [0,0.01],ax = ax[0][0],autopct="%.1f%%")
ax[0][0].set_title("스트레스를 받은 적이 있다")
ax[0][0].set_ylabel('')

ori["우울감경험율"].astype(float).plot.pie(explode = [0,0.01],ax = ax[0][1],autopct="%.1f%%")
ax[0][1].set_title("우울증을 경험한 적이 있다")
ax[0][1].set_ylabel('')

ori["자살생각율"].astype(float).plot.pie(explode = [0,0.01],ax = ax[0][2],autopct="%.1f%%")
ax[0][2].set_title("자살을 고민한 적이 있다.")
ax[0][2].set_ylabel('')


ori2["스트레스"].astype(float).plot.pie(explode = [0,0.01],ax = ax[1][0],autopct="%.1f%%")
ax[1][0].set_title("스트레스를 받은 적이 있다")
ax[1][0].set_ylabel('')

ori2["우울감경험율"].astype(float).plot.pie(explode = [0,0.01],ax = ax[1][1],autopct="%.1f%%")
ax[1][1].set_title("우울증을 경혐한 적이 있다")
ax[1][1].set_ylabel('')

ori2["자살생각율"].astype(float).plot.pie(explode = [0,0.01],ax = ax[1][2],autopct="%.1f%%")
ax[1][2].set_title("자살을 고민한 적이 있다")
ax[1][2].set_ylabel('')

plt.suptitle('2024~2025 학생 정신 건강 관련 통계',fontsize = 16)
plt.show()
```
